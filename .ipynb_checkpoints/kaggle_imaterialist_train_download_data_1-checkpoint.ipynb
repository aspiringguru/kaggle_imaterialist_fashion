```python
import imghdr
import urllib.request
import shutil
import os
import time
import sys
```

```python
DATADIR = "/home/ec2-user/kaggleData/imaterialist/"
TEST_DIR = DATADIR + "test"
TRAIN_DIR = DATADIR + "train"
VALID_DIR = DATADIR + "valid"
KAGGLE_DATA_DIR = "/home/ec2-user/.kaggle/competitions/imaterialist-challenge-fashion-2018/"
TEST_SOURCE = "test.json"
TRAIN_SOURCE = "train.json"
VALIDATION_SOURCE = "validation.json"
```

```python
import json

with open(KAGGLE_DATA_DIR+TRAIN_SOURCE, 'r') as f:
    train_source = json.load(f)

print (type(train_source), len(train_source))
```

```python
print ("train_source.keys():", train_source.keys())
print ("train_source['images']:", type(train_source['images']), len(train_source['images']))
print (type(train_source['images'][0]), train_source['images'][0].keys() )
```

```python
#list files in dir  for file in 
trainFilesList = os.listdir(TRAIN_DIR)
print ("No of files :", len(trainFilesList))
for i in range(len(trainFilesList)):
    temp = trainFilesList[i].split(".",1)[0]  #strip .filesuffix and convert to int.
    print ("i:", i, ":", trainFilesList[i], temp)
    trainFilesList[i] = int(temp)
```

```python
type(trainFilesList[0])
```

```python
os.chdir(TRAIN_DIR)

err_count = 0
start_time = time.time()
for image in train_source['images']:
    try:
        print (image['imageId'], image['url'])
        if image['imageId'] not in trainFilesList:
            # Download the file from `url` and save it locally under `file_name`:
            with urllib.request.urlopen(image['url']) as response, open(str(image['imageId']), 'wb') as out_file:
                shutil.copyfileobj(response, out_file)
            image_type = imghdr.what(str(image['imageId']))
            print(image_type)
            os.rename(str(image['imageId']), str(image['imageId'])+"."+image_type)
        else:
            print ("already downloaded, skipping.")
    except:
        print ("Unexpected error: "+ image['imageId'], image['url'], sys.exc_info()[0])
        err_count += 1
    #break
print ("err_count:", err_count)
print ("execution took :"+str(time.time()-start_time))
```
